```python
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
```

```python
df.shape
```

```python
df.head()
```

```python
plt.plot(df['year'], df['per capita income (US$)'])
```

```python
x = df.iloc[:37,0].values
y = df.iloc[:37,1].values
x = x.reshape(-1,1)
y = y.reshape(-1,1)
y
```

```python
reg = linear_model.LinearRegression().fit(x,y)
```

```python
x_t = df.iloc[37:, 0].values
x_t = x.reshape(-1,1)
y_t = df.iloc[37:, 1].values
y_t = y.reshape(-1,1)
```

```python
pred = reg.predict(x_t)
```

```python
plt.plot(x_t,y_t, color = 'r')
plt.plot(x_t,pred, color='b')
plt.xlabel('year')
plt.ylabel('price')
```
